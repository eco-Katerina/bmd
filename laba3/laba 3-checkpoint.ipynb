```python
import seaborn as sns
%matplotlib inline
```

Збереження даних:

```python
tips = sns.load_dataset('tips')
```

```python
tips.head
```

Візуалізація даних:

```python
sns.distplot(tips['total_bill'],color="y")
```

```python
sns.jointplot(x='total_bill', y='tip', data = tips, color="y")
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

Графіки для аналізу категоріальних даних:

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips)
```

```python
import numpy as np
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.median)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, palette = 'coolwarm')
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips )
```
