Створіть вектор розміром 11 зі значеннями від 0 до 5. Розрахуйте на його основі іектор у (y = x^2).

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

Побудова графіків.
1. Функціональний метод.

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
```

2. Об’єктно-орієнтований метод.


```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

Побудова двох графіків в одній області

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])


axes1.plot(x,y,'r')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Title')

axes2.plot(x, y, marker = 'p', markersize=15, markerfacecolor = 'y')
axes2.set_xlabel('Y Label')
axes2.set_ylabel('X Label')
axes2.set_title('Title')
```

```python
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,color='b')
axes[1].plot(y,x,color='g')
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2",lw=4)
ax.plot(x, x**3, label="x**3",ls='-.')
ax.legend()
```
